```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('product_cat_subcat_mapping.csv')
df2 = pd.read_csv('product_level_details.csv')
```

```python
df1.head(2)
```

```python
df2.head(2)
```

```python
df = pd.merge(df1, df2, on='product_id')
df.head(2)
```

```python
# 1:- Category wise sales and orders on Day-on-Day basis
df.groupby(['order_date','product_category_id'])['sales'].aggregate(['sum'])
```

```python
#2 Top 10 categories in terms of orders on month on month basis

df['month'] = pd.to_datetime(df['order_date']).dt.strftime('%b')
df.head(2)
```

```python
ddf = df.groupby(['month','product_category_id'])['order_id'].aggregate(['count'])
ddf.groupby(['month'])['count'].nlargest(10).head()
```

```python
3# First order date of every customer in each month
df.groupby(['month','customer_id'])['order_date'].aggregate(['min'])
```
